## Análisis de componentes de principales - Principal Component Analysis (PCA)

![gifpca](https://miro.medium.com/v2/resize:fit:1400/format:webp/0*TXIbCytfgo5p5hty.gif)


- PCA busca un espacio de menor dimensión que preserve la mayor cantidad de variabilidad posible desde el espacio original (alta dimensión).

- Desde una perspectiva generativa, PCA asume que los datos $\mathbf{X} \in \mathbb{R}^{N \times P}$ son generados desde un espacio de baja dimensión (espacio latente) $\mathbf{Z} \in \mathbb{R}^{N \times M}$, donde $M < P$:

\begin{equation}
\mathbf{X} = \mathbf{Z} \mathbf{W}^T,
\end{equation}

- $\mathbf{Z}=\mathbf{X}\mathbf{W}$: matriz de baja dimensión.
- $\mathbf{W} \in \mathbb{R}^{P \times M}$: componentes principales (loading vectors).

El problema de optimización desde el punto de vista generativo:

\begin{align}
\mathbf{W}^*=&\arg\min_\mathbf{W} \|\mathbf{X}-\mathbf{Z} \mathbf{W}^T\|_F \\
   \text{s.t.} & \quad \mathbf{W}\mathbf{W}^\top  = \mathbf{I}
\end{align}



- PCA también puede resolverse maximizando la varianza de los datos proyectados.

- Asumiendo datos centralizados $\mathbf{X}$, la covarianza $\mathbf{\Sigma}\in \mathbb{R}^{P \times P}$ se define como:

\begin{equation}
\mathbf{\Sigma} = \frac{1}{N} \mathbf{X}^T \mathbf{X}.
\end{equation}

- El problema de optimización de PCA se puede reescribir como:

\begin{align}
\mathbf{W}^*=&\arg\max_\mathbf{W} \text{tr}(\mathbf{W}^T \mathbf{\Sigma} \mathbf{W}),\\
 \text{s.t.} & \quad \mathbf{W}\mathbf{W}^\top  = \mathbf{I}
\end{align}


- La solución del problema se centra en los $M$ vectores propios de la matriz de covarianza, correspodiente a los $M$ valores propios más altos.

- Ver la demostración [PCA generativo a PCA por varianza](https://github.com/amalvarezme/AprendizajeMaquina/blob/main/4_Clasificacion_Clustering_DR/otros/DemostracionPCA.pdf).


**Ejemplo:** Construyamos un conjunto de datos 3D del famoso dato sintético rollo Suizo - Swiss Roll.

```python
from sklearn.datasets import make_swiss_roll
import numpy as np
import matplotlib.pyplot as plt

X, t = make_swiss_roll(n_samples=1000, noise=0.2, random_state=42)
#t contiene el orden estructural de los datos
#crear scatter en 3D
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
ax.scatter3D(X[:,0], X[:,1], X[:,2],c=t)
ax.view_init(elev=10, azim=-75)
plt.show()
```

**NOTA:** PCA supone que el conjunto de datos se centra en el origen. Como veremos, las clases de PCA de Scikit-Learn se encargan de centrar los datos por usted. Sin embargo, si implementa PCA usted mismo, o si usa otras bibliotecas, **no olvide centrar los datos primero**, incluso considere aplicar normalización zscore.

**Usando PCA de scikit-learn**

-La clase PCA de Scikit-Learn implementa PCA usando la descomposición espectral (EIG y SVD).

-El siguiente código aplica PCA para reducir la dimensionalidad del conjunto de datos a dos dimensiones (tenga en cuenta que automáticamente se encarga de centrar los datos):

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 3)
Y2D = pca.fit_transform(X)
```

- El hiperparámetro `n_components` puede ser un número entero (número de dimensiones o componentes $M\leq P$) o un valor en el intervalo [0,1] para cuantificar el número de componentes requerido para preservar una varianza retenida dada.

Tenga en cuenta que ejecutar PCA varias veces en conjuntos de datos ligeramente diferentes puede generar espacios latentes distintos. En general, la única diferencia es que algunos ejes pueden voltearse.

**Nota**: el problema de optimización de PCA que incluye costo y restricciones cuadráticas, da como resultado un problema no convexo.

```python
plt.scatter(Y2D[:,0],Y2D[:,1],c=t)
plt.xlabel('componente 1')
plt.ylabel('componente 2')
plt.show()
```

Ahora vamos a recuperar los puntos 3D proyectados en el plano (subespacio PCA 2D).

```python
X3D_inv = pca.inverse_transform(Y2D)
```

Por supuesto, **hubo cierta pérdida de información durante el paso de proyección**, por lo que los puntos 3D recuperados *no son exactamente iguales a los puntos 3D originales*:

```python
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
ax.scatter3D(X3D_inv[:,0], X3D_inv[:,1], X3D_inv[:,2],c=t)
ax.view_init(elev=10, azim=-75)
plt.show()
```

Podemos calcular el error de reconstrucción:

```python
np.mean(np.sum(np.square(X3D_inv - X), axis=1))
```

Las reconstrucciones de ambos métodos **no son idénticas** porque la clase PCA de Scikit-Learn se encarga automáticamente de revertir el centrado medio, *pero si restamos la media, obtenemos la misma reconstrucción*.

## Ejercicio: PCA – Formulación generativa vs. preservación de la varianza

Dado un conjunto de datos centrado en la media $\mathbb{E}[\mathbf{x}] = 0$, donde $\mathbf{x} \in \mathbb{R}^p$, buscamos encontrar una dirección $\mathbf{w} \in \mathbb{R}^p$, con $\|\mathbf{w}\| = 1$, que permita proyectar los datos en una dimensión:

$$
z = \mathbf{w}^\top \mathbf{x} \in \mathbb{R}, \quad \hat{\mathbf{x}} = \mathbf{w} z = \mathbf{w} \mathbf{w}^\top \mathbf{x}
$$

Se nos pide:

1. Presentar PCA como un algoritmo **generativo**, es decir, que reconstruye los datos proyectados.
2. Comparar esta formulación con el enfoque de **preservación de varianza**.

---

### 1. Formulación generativa – Minimización del error de reconstrucción

La reconstrucción de $\mathbf{x}$ a partir de $z$ es:

$$
\hat{\mathbf{x}} = \mathbf{w} \mathbf{w}^\top \mathbf{x}
$$

Queremos minimizar el error cuadrático medio de reconstrucción:

$$
\mathbf{w}^* = \arg \min_{\|\mathbf{w}\| = 1} \mathbb{E} \left[ \| \mathbf{x} - \hat{\mathbf{x}} \|^2 \right]
$$

Expandimos el error usando la norma de Frobenius:

$$
\| \mathbf{x} - \hat{\mathbf{x}} \|^2 = \| \mathbf{x} - \mathbf{w} \mathbf{w}^\top \mathbf{x} \|^2 = \mathbf{x}^\top \mathbf{x} - 2 (\mathbf{w}^\top \mathbf{x})^2 + (\mathbf{w}^\top \mathbf{x})^2 = \mathbf{x}^\top \mathbf{x} - (\mathbf{w}^\top \mathbf{x})^2
$$

Tomando esperanza:

$$
\mathbb{E} \left[ \| \mathbf{x} - \hat{\mathbf{x}} \|^2 \right] = \mathbb{E} \left[ \mathbf{x}^\top \mathbf{x} \right] - \mathbb{E} \left[ (\mathbf{w}^\top \mathbf{x})^2 \right]
$$

El primer término no depende de $\mathbf{w}$, por lo tanto minimizar el error equivale a:

$$
\max_{\|\mathbf{w}\| = 1} \mathbb{E} \left[ (\mathbf{w}^\top \mathbf{x})^2 \right]
$$

---

### 2. Formulación de máxima varianza

La matriz de covarianza del conjunto es:

$$
\Sigma = \mathbb{E}[\mathbf{x} \mathbf{x}^\top] \in \mathbb{R}^{p \times p}
$$

Entonces,

$$
\mathbb{E} \left[ (\mathbf{w}^\top \mathbf{x})^2 \right] = \mathbf{w}^\top \mathbb{E}[\mathbf{x} \mathbf{x}^\top] \mathbf{w} = \mathbf{w}^\top \Sigma \mathbf{w}
$$

Así, el problema se convierte en:

$$
\mathbf{w}^* = \arg \max_{\|\mathbf{w}\| = 1} \mathbf{w}^\top \Sigma \mathbf{w}
$$

Este es el enfoque de **máxima varianza**.

---

### 3. Solución mediante valores propios

Usamos multiplicadores de Lagrange para incorporar la restricción $\|\mathbf{w}\| = 1$:

$$
\mathcal{L}(\mathbf{w}, \lambda) = \mathbf{w}^\top \Sigma \mathbf{w} - \lambda (\mathbf{w}^\top \mathbf{w} - 1)
$$

Derivando:

$$
\frac{\partial \mathcal{L}}{\partial \mathbf{w}} = 2 \Sigma \mathbf{w} - 2 \lambda \mathbf{w} = 0 \quad \Rightarrow \quad \Sigma \mathbf{w} = \lambda \mathbf{w}
$$

Por tanto, $\mathbf{w}$ debe ser un **vector propio de $\Sigma$**, y la solución óptima es el vector propio asociado al **mayor valor propio $\lambda_{\max}$**.

---

### 4. Comparación de las formulaciones

| Perspectiva               | Función objetivo                                  | Intuición                                                                 | Solución                          |
|--------------------------|---------------------------------------------------|--------------------------------------------------------------------------|-----------------------------------|
| Generativa / reconstrucción | $\min \mathbb{E} \left[ \| \mathbf{x} - \hat{\mathbf{x}} \|^2 \right]$ | Escoge el subespacio que **mejor reconstruye** los datos                | Llega al problema de autovalores |
| Preservar varianza        | $\max \mathbf{w}^\top \Sigma \mathbf{w}$      | Escoge el subespacio que **captura mayor dispersión** de los datos      | Directamente problema de autovalores |

Ambas formulaciones conducen al mismo problema de valores propios. Si se desean más componentes (dimensión reducida $M$), se seleccionan los $M$ vectores propios principales de $\Sigma$:

$$
\mathbf{W}^* = [\mathbf{e}_1 \; \mathbf{e}_2 \; \dots \; \mathbf{e}_M], \quad \Sigma \mathbf{e}_i = \lambda_i \mathbf{e}_i, \quad \lambda_1 \geq \lambda_2 \geq \dots \geq \lambda_M
$$

---

### ✅ Conclusión

Minimizar el error de reconstrucción (formulación generativa) y maximizar la varianza (formulación estadística) son **dos caminos equivalentes** que conducen al mismo resultado en PCA. Ambos se resuelven obteniendo los vectores propios principales de la matriz de covarianza del conjunto de datos.

**Ejemplo:** En el siguiente código se comprime el conjunto de datos MNIST a $154$ dimensiones, luego usa el método *inverse_transform()* para descomprimirlo nuevamente a $784$ dimensiones. La  siguiente Figura muestra algunos dígitos del conjunto de entrenamiento original (a la izquierda) y los dígitos correspondientes después de la compresión y descompresión. Puede ver que hay una ligera pérdida de calidad de imagen, pero los dígitos aún están casi intactos.

```python
import matplotlib as mpl

pca         = PCA(n_components = 154)
X_reduced   = pca.fit_transform(X_train)
X_recovered = pca.inverse_transform(X_reduced)

def plot_digits(instances, images_per_row=5, **options):
    size           = 28
    images_per_row = min(len(instances), images_per_row)
    print(instances.shape)
    images         = [instance.reshape(size,size) for instance in instances]
    n_rows         = (len(instances) - 1) // images_per_row + 1
    row_images     = []
    n_empty        = n_rows * images_per_row - len(instances)
    images.append(np.zeros((size, size * n_empty)))
    for row in range(n_rows):
        rimages = images[row * images_per_row : (row + 1) * images_per_row]
        row_images.append(np.concatenate(rimages, axis=1))
    image = np.concatenate(row_images, axis=0)
    plt.imshow(image, cmap = mpl.cm.binary, **options)
    plt.axis("off")

plt.figure(figsize=(7, 4))
plt.subplot(121)
plot_digits(X_train[::40])
plt.title("Original", fontsize=16)
plt.subplot(122)
plot_digits(X_recovered[::40])
plt.title("Compressed", fontsize=16)
#save_fig("mnist_compression_plot")
```

### Kernel PCA

- El algoritmo Kernel PCA permite calcular los componentes principales a partir de un mapeo no lineal basado en funciones núcleo.

- En KPCA los datos en $\mathbf{X}$ son mapeados a un espacio de alta dimensión desde la función $\phi: \mathbb{R}^P \to \mathscr{F}$.

- Luego, se encuentran los componentes principales sobre la matriz kernel $\mathbf{K}\in \mathbb{R}^{N \times N}$, con $K_{nn'} = \kappa(\mathbf{x}_n,\mathbf{x}_{n'})=\langle \phi(\mathbf{x}_n),\phi(\mathbf{x}_{n'})\rangle$.


\begin{equation}
\mathbf{K} \boldsymbol{\alpha} = \lambda \boldsymbol{\alpha},
\end{equation}


- Para encontrar la proyección de los datos:

\begin{equation}
\mathbf{z}_n = \sum_{n'=1}^N \alpha_{n'} \kappa(\mathbf{x}_n, \mathbf{x}_{n'}), \quad n = 1, 2, \ldots, N,
\end{equation}

### Ejercicio – Paralelo entre PCA lineal y Kernel PCA

---

#### 1 · Notación  

- Datos centrados  
  $$\mathbf X\in\mathbb R^{N\times P},\qquad \sum_{n=1}^{N}\mathbf x_n=\mathbf 0.$$
- En Kernel PCA (KPCA) existe un mapeo no lineal  
  $$\boldsymbol\phi:\mathbb R^{P}\longrightarrow\mathscr F,$$  
  y un núcleo  
  $$\kappa(\mathbf x,\mathbf x')=\langle\boldsymbol\phi(\mathbf x),\boldsymbol\phi(\mathbf x')\rangle.$$

---

#### 2 · Objetivo de PCA lineal  

$$
\boxed{%
\max_{\mathbf W\in\mathbb R^{P\times M}}
\operatorname{tr}\!\bigl(\mathbf W^{\!\top}\mathbf\Sigma\mathbf W\bigr)
\quad
\text{s. a. }\mathbf W^{\!\top}\mathbf W=\mathbf I_M
}
\tag{P1}
$$  

donde  

$$
\mathbf\Sigma=\frac1N\,\mathbf X^{\!\top}\mathbf X.
$$  

La solución se obtiene tomando los $M$ vectores propios principales de $\mathbf\Sigma$.

---

#### 3 · Objetivo de Kernel PCA  

En el espacio de características:

$$
\mathbf C
=\frac1N\sum_{n=1}^{N}
  \bigl(\boldsymbol\phi_n-\bar{\boldsymbol\phi}\bigr)
  \bigl(\boldsymbol\phi_n-\bar{\boldsymbol\phi}\bigr)^{\!\top},
\qquad
\bar{\boldsymbol\phi}=\frac1N\sum_{n=1}^{N}\boldsymbol\phi_n.
$$

Buscamos $\mathbf v\in\mathscr F,\;\lVert\mathbf v\rVert=1$ tal que  

$$
\max_{\lVert\mathbf v\rVert=1}\,
\mathbf v^{\!\top}\mathbf C\,\mathbf v
\quad\Longrightarrow\quad
\mathbf C\mathbf v=\lambda\mathbf v.
$$

**Teorema del representer**  
Todo vector propio puede escribirse como  

$$
\boxed{\;
\mathbf v=\sum_{n=1}^{N}\alpha_n\,
\bigl(\boldsymbol\phi_n-\bar{\boldsymbol\phi}\bigr)
\;}
$$

y conduce al problema matricial  

$$
\boxed{%
\mathbf K_{\!\mathrm c}\boldsymbol\alpha
  =N\lambda\,\boldsymbol\alpha,
\qquad
\boldsymbol\alpha^{\!\top}\mathbf K_{\!\mathrm c}\boldsymbol\alpha=1
}
\tag{KP1}
$$

donde  

$$
\mathbf K_{ij}=\kappa(\mathbf x_i,\mathbf x_j),\qquad
\mathbf K_{\!\mathrm c}=
\mathbf H\mathbf K\mathbf H,\qquad
\mathbf H=\mathbf I_N-\tfrac1N\mathbf 1\mathbf 1^{\!\top}.
$$

---

#### 4 · Proyección de un punto nuevo  

| Método | Coordenada sobre el componente \(m\) |
|--------|--------------------------------------|
| **PCA** | $z_m=\mathbf e_m^{\!\top}\mathbf x$ |
| **KPCA** | $z_m=\displaystyle\sum_{n=1}^{N}\alpha_n^{(m)}\,\kappa(\mathbf x,\mathbf x_n)$ |

---

#### 5 · Comparación resumida  

| Aspecto | PCA lineal | Kernel PCA |
|---------|------------|------------|
| Espacio donde se optimiza | $\mathbb R^{P}$ | $\mathscr F$ |
| Matriz a diagonalizar | $\mathbf\Sigma=\tfrac1N\mathbf X^{\!\top}\mathbf X$ | $\mathbf K_{\!\mathrm c}$ |
| Variable propia | $\mathbf e\in\mathbb R^{P}$ | $\boldsymbol\alpha\in\mathbb R^{N}$ |
| Restricción | $\mathbf e^{\!\top}\mathbf e=1$ | $\boldsymbol\alpha^{\!\top}\mathbf K_{\!\mathrm c}\boldsymbol\alpha=1$ |
| Captura no linealidad | ✗ | ✓ |
| Reconstrucción (*pre-image*) | Explícita: $\hat{\mathbf x}=z\mathbf e+\bar{\mathbf x}$ | No cerrada; suele requerir optimización |

---

#### 6 · Conclusión  

Ambos métodos se formulan como un **problema de autovalores con restricción de ortonormalidad**.  
La diferencia es **dónde** se mide la varianza:  
- PCA lineal lo hace en el espacio original $\mathbb R^{P}$.  
- Kernel PCA la mide en un espacio de características $\mathscr F$ accesible mediante el núcleo, lo que permite capturar **patrones no lineales** (por ejemplo, desenrollar el *Swiss Roll*) sin abandonar el marco algebraico de PCA.

**Ejemplo:** el siguiente código utiliza la clase *KernelPCA* de Scikit-Learn para realizar kPCA con un kernel RBF.

### kPCA: Definiendo el kernel y sus hiperparámetros

- Como kPCA es un algoritmo de aprendizaje no supervisado, *no existe una medida de rendimiento obvia para ayudarlo a seleccionar los mejores valores de kernel e hiperparámetros*.

- Sin embargo, la reducción de dimensionalidad es a menudo un paso de preparación para una tarea de aprendizaje supervisada (por ejemplo, clasificación), por lo que simplemente puede usar la **búsqueda de cuadrícula** para seleccionar el núcleo y los hiperparámetros conducen al mejor rendimiento en esa tarea.

**Ejemplo:** el siguiente código crea una guía de dos pasos para la clasificación de la base de datos Mnist.

- Primero reduciendo la dimensionalidad a dos dimensiones usando kPCA, luego aplicando Regresión logística para la clasificación.

- Se utiliza *GridSearchCV* para encontrar el mejor valor de kernel y gamma para kPCA para obtener la mejor precisión de clasificación al final de la guía:

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

clf = Pipeline([
        ("kpca", KernelPCA(n_components=2)),
        ("log_reg", LogisticRegression(solver="lbfgs"))
    ])

param_grid = [{
        "kpca__gamma": np.linspace(0.03, 0.05, 10),
        "kpca__kernel": ["rbf", "sigmoid"]
    }]

grid_search = GridSearchCV(clf, param_grid, cv=3)
grid_search.fit(X_train, y_train)
```

Los mejores kernels e hiperparámetros están disponibles a través de la variable *best_params_*:

```python
print(grid_search.best_params_)
```

- Una solución es entrenar un modelo de regresión supervisado, con las instancias proyectadas como el conjunto de entrenamiento y las instancias originales como los objetivos.

- Scikit-Learn lo hará automáticamente si configura *fit_inverse_transform = True*, ([pre-image problem](https://www.aaai.org/Papers/ICML/2003/ICML03-055.pdf)) como se muestra en el siguiente código:

```python
from sklearn.metrics import mean_squared_error

rbf_pca    = KernelPCA(n_components = 2, kernel="rbf", gamma=0.0433,
                    fit_inverse_transform=True)
X_reduced  = rbf_pca.fit_transform(X)
X_preimage = rbf_pca.inverse_transform(X_reduced)
mean_squared_error(X, X_preimage)
```

## t-distributed stochastic neighborhood embedding

- [t-SNE](https://www.jmlr.org/papers/volume9/vandermaaten08a/vandermaaten08a.pdf?fbclid=IwA) es de los algoritmos más potentes en reducción de dimensión dado que permite unir los principios de manifold learning y probabilidades, para lidear con el problema de concentración de las normas (métricas en espacios de alta dimensión se distribuyen muy diferente a métricas en espacios de baja dimensión).


![gifpca](https://learnopencv.com/wp-content/uploads/2022/11/tsne.gif)



- t-SNE busca minimizar la diferencia entre las funciones de probabilidad del espacio original (alta dimensión) y el espacio reducido (baja dimensión), evitando el problema de la concentración de la norma.


- La similitud es modelada como una distribución Gaussiana, tal que la probabilidad condicional $p_{n'|n}$ que define si  $\mathbf{x}_n$ selecciona a $\mathbf{x}_{n'}$ como su vecino es:

\begin{equation}
p_{n'|n} = \frac{\exp(-\|\mathbf{x}_n - \mathbf{x}_{n'}\|^2 / 2\sigma_n^2)}{\sum_{k \neq n} \exp(-\|\mathbf{x}_n - \mathbf{x}_k\|^2 / 2\sigma_n^2)},
\end{equation}

donde $\sigma_n\in \mathbb{R}^+$.

La probabilidad conjunta se puede definir como :

\begin{equation}
p_{nn'} = \frac{p_{n'|n} + p_{n|n'}}{2N}.
\end{equation}



- Para el espacio de baja dimensión se utiliza un distribución t-Student con un grado de libertad (distribución de Cauchy).

- La probabilidad conjunta $q_{nn'}$ que modela si $\mathbf{y}_n$ y $\mathbf{z}_{n'}$ están cercanos es dada por:

\begin{equation}
q_{nn'} = \frac{(1 + \|\mathbf{z}_n - \mathbf{z}_{n'}\|^2)^{-1}}{\sum_{k \neq l} (1 + \|\mathbf{z}_k - \mathbf{z}_l\|^2)^{-1}}.
\end{equation}


- t-SNE minimiza la divergencia de  Kullback-Leibler (KL) entre $P = \{p_{nn'}\}$ y $Q = \{q_{nn'}\}$:

\begin{equation}
\text{KL}(P \| Q) = \sum_{n \neq n'} p_{nn'} \log \left( \frac{p_{nn'}}{q_{nn'}} \right).
\end{equation}

- La minización se resuelve convencionalmente por métodos de gradiente.

- Es importante resaltar que el ancho de bando en el espacio de alta dimensión se fija por muestra, fijando un valor de entropia común (perplexity).

## Uniform Manifold Approximation and Projection (UMAP)

![gifumap](https://umap-learn.readthedocs.io/en/latest/_images/aligned_umap_pendigits_anim.gif)

- UMAP es una técnica de reducción de dimensión que modela el espacio de alta dimensión como estructura topológica difusa y la optimiza en el espacio de baja dimensión preservando distancias globales y locales.

- UMAP construye un grafo ponderado para representar la estructura topológica difusa.

- La probabilidad $p_{nn'}$ que representa el peso del borde entre $\mathbf{x}_n$ and $\mathbf{x}_{n'}$ esta dada por:

\begin{equation}
p_{nn'} = \exp\left(-\frac{\|\mathbf{x}_n - \mathbf{x}_{n'}\| - \rho_n}{\sigma_n}\right),
\end{equation}

- $\rho_n$ es la distancia al vecino más cercano de $\mathbf{x}_n$, lo que asegura que las distancias locales no se ven afectadas por ruido.
    
- $\sigma_n$ se determina por una un número fijo de vecinos.


- La relación difusa se vuelve simétrica como:

\begin{equation}
\tilde{p}_{nn'} = p_{nn'} + p_{n'n} - p_{nn'} p_{n'n}.
\end{equation}



- Las relaciones en el espacio de baja dimensión se fijan mediante una distribución t-student de cola pesada:

\begin{equation}
q_{nn'} = \left(1 + a \|\mathbf{z}_n - \mathbf{z}_{n'}\|^{2b}\right)^{-1},
\end{equation}

con $a$ y $b$ como parámetros de la distribución. Generalmente se fijan en 1.


- UMAP minimiza la entropía cruzada :

\begin{equation}
C(P;Q) = \sum_{n \neq n'} \left( -\tilde{p}_{nn'} \log(q_{nn'}) - (1 - \tilde{p}_{nn'}) \log(1 - q_{nn'}) \right).
\end{equation}

- Convencionalmente se utilizan métodos de gradiente para resolver la optimización.

# Fundamento teórico de Eigenfaces usando PCA
# -----------------------------------------

"""
PCA transforma imágenes de alta dimensión (ej: 784D para 28x28 píxeles) a un
subespacio de menor dimensión que capture la máxima **varianza**.
"""

# Formulación Matemática
"""
Dado un conjunto de imágenes vectorizadas:
$$ X = \begin{bmatrix} x_1 \\ x_2 \\ \vdots \\ x_N \end{bmatrix} \in \mathbb{R}^{N \times P} $$
donde cada $x_i \in \mathbb{R}^P$ (P = 784 para imágenes 28x28).
"""

"""
1. **Centrado de datos:**
$$ X_c = X - 1_N \mu^T, \quad \text{con } \mu = \frac{1}{N} \sum_{i=1}^{N} x_i $$
"""

"""
2. **Matriz de covarianza empírica:**
$$ \Sigma = \frac{1}{N} X_c^T X_c \in \mathbb{R}^{P \times P} $$
"""

"""
3. **Descomposición espectral (autovectores/autovalores):**
Se calculan los autovectores $w_1, \ldots, w_M$ para los $M$ mayores autovalores $\lambda_1 \geq \lambda_2 \geq \ldots$:
$$ \Sigma w_i = \lambda_i w_i $$
"""

"""
4. **Proyección al subespacio (Eigenfaces):**
$$ Z = X_c W \quad \text{donde } W = [w_1, \ldots, w_M] $$
"""

"""
5. **Reconstrucción aproximada:**
$$ \hat{X} = ZW^T + \mu $$
"""

# Implementación práctica en Python (opcional):
"""

pca = PCA(n_components=M)  # M = dimensiones deseadas
Z = pca.fit_transform(X_c)  # Proyección
X_reconstruido = pca.inverse_transform(Z) + mu  # Reconstrucción
"""

```python
# Importar librerías necesarias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
# No es necesario importar mnist desde tensorflow if using fetch_openml
# from tensorflow.keras.datasets import mnist
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
from sklearn.manifold import TSNE
from sklearn.decomposition import KernelPCA
import umap
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import seaborn as sns
from sklearn.datasets import fetch_openml # Importar fetch_openml

# Cargar MNIST
mnist = fetch_openml('mnist_784', version=1)
X = mnist.data / 255.0
y = mnist.target.astype(int)

# Submuestreo
X_train = X[:500]
y_train = y[:500]

# PCA
pca = PCA(n_components=0.95)
X_pca = pca.fit_transform(X_train)

# Proyección 2D
X_2d = X_pca[:, :2]

# Graficar puntos con imágenes
def plot_mnist_2d(Z, images, img_w=28, img_h=28, zoom=0.5, step=20):
    fig, ax = plt.subplots(figsize=(12, 10))
    for i in range(0, len(images), step):  # solo cada `step` imágenes
        # Use .iloc for integer-based indexing on the DataFrame
        image = images.iloc[i].values.reshape((img_w, img_h))
        im = OffsetImage(image, zoom=zoom, cmap='gray')
        ab = AnnotationBbox(im, (Z[i, 0], Z[i, 1]), frameon=False)
        ax.add_artist(ab)
    ax.set_title("Proyección PCA 2D con imágenes")
    ax.set_xlim(Z[:,0].min() - 1, Z[:,0].max() + 1)
    ax.set_ylim(Z[:,1].min() - 1, Z[:,1].max() + 1)
    plt.grid(True)
    plt.show()

plot_mnist_2d(X_2d, X_train, step=20)

# Mostrar las cinco primeras eigenfaces
fig, axs = plt.subplots(1, 5, figsize=(15, 3))
for i in range(5):
    axs[i].imshow(pca.components_[i].reshape(28, 28), cmap="gray")
    axs[i].set_title(f"Eigenface {i+1}")
    axs[i].axis('off')
plt.suptitle("Primeras 5 Eigenfaces")
plt.show()

# Parte 2: Comparación de técnicas con coeficiente de silueta
np.random.seed(42)
idx = np.random.choice(X.shape[0], 3000, replace=False)
X_sample = X.iloc[idx] # Usar .iloc para seleccionar filas del DataFrame
y_sample = y.iloc[idx] # Usar .iloc para seleccionar filas de la Serie

# Reducir a 2D con diferentes técnicas
proj = {}
proj["PCA"] = PCA(n_components=2).fit_transform(X_sample)
proj["Kernel PCA"] = KernelPCA(n_components=2, kernel="rbf", gamma=0.04).fit_transform(X_sample)
# Asegúrate de que los datos de entrada para t-SNE y UMAP sean numpy arrays
proj["t-SNE"] = TSNE(n_components=2, perplexity=30, learning_rate=200, random_state=42).fit_transform(X_sample.values)
proj["UMAP"] = umap.UMAP(n_components=2, random_state=42).fit_transform(X_sample.values)

# Calcular silueta con clustering
scores = {}
labels_dict = {}
for method, Z in proj.items():
    # Ensure Z is a numpy array for KMeans
    kmeans = KMeans(n_clusters=10, random_state=42, n_init=10) # Add n_init to suppress warning
    labels = kmeans.fit_predict(Z)
    labels_dict[method] = labels
    score = silhouette_score(Z, labels)
    scores[method] = score

# Graficar resultados de reducción con color según clustering
fig, axs = plt.subplots(2, 2, figsize=(14, 10))
axs = axs.flatten()
for i, (method, Z) in enumerate(proj.items()):
    axs[i].scatter(Z[:, 0], Z[:, 1], c=labels_dict[method], cmap="tab10", s=10)
    axs[i].set_title(f"{method} (Silhouette = {scores[method]:.4f})")
    axs[i].grid(True)
plt.tight_layout()
plt.show()

# Mostrar coeficientes de silueta en gráfico de barras
plt.figure(figsize=(8, 4))
sns.barplot(x=list(scores.keys()), y=list(scores.values()))
plt.ylabel("Coeficiente de silueta")
plt.title("Comparación de métodos de reducción de dimensión")
plt.ylim(0, 1)
plt.grid(True, axis='y')
plt.show()
```
